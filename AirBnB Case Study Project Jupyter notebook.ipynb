```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
air_bnb = pd.read_csv('./AB_NYC_2019.csv')
air_bnb
```

```python
# How many neighborhood groups are available and which shows up the most?
air_bnb_neigh = air_bnb.groupby(['neighbourhood_group']).count()
air_bnb_neigh

# There are 5 different neighbourhood groups available and the one that shows up the most is Manhattan with
# over 21 thousand entries
```

```python
# Are private rooms the most popular in manhattan?

air_bnb_roomtypes= air_bnb.groupby(['room_type', 'neighbourhood_group']).count()
air_bnb_roomtypes

# The 'Private Room' option is most certainly one of the popular available choices in Manhattan with 7982 entries,
# but the 'Entire home/apt' room type has almost double the offerings => 13199.
```

```python
# Which hosts are the busiest based on their reviews?
air_bnb_busyhost= air_bnb.groupby(['host_id', 'host_name', 'calculated_host_listings_count'], as_index=False)['number_of_reviews'].sum()
air_bnb_busyhost.sort_values('number_of_reviews', ascending=False).head()

# This data shows the total number of reviews between all the listings/properties a host has. 
# This shows that Maya has the most reviews between her 5 listings followed by Len with his 13 listings and 
# his 2205 total reviews
```

```python
# Which hosts are the busiest based on their reviews?
air_bnb_busyhost= air_bnb.groupby(['host_id', 'host_name', 'calculated_host_listings_count'], as_index=False)['reviews_per_month'].sum()
air_bnb_busyhost.sort_values('reviews_per_month', ascending=False).head()

# If we analyse data based on a monthly number of reviews between all listings of a certain host, Sonder has a 
# total of 397.56 reviews per month followed by Row with 111.72
```

```python
#Which neighorhood group has the highest average price?
air_bnb_avg_price = air_bnb.groupby(['neighbourhood_group'], as_index=False)['price'].mean()
air_bnb_avg_price.sort_values('price', ascending=False)

# The highest average price for a room in NYC is in Manhattan with a little over 196 $/night
```

```python
# Which neighbor hood group has the highest total price?
air_bnb_total_price = air_bnb.groupby(['neighbourhood_group'], as_index=False)['price'].sum()
air_bnb_total_price.sort_values('price', ascending=False)
# The highest total price for listings in NYC are in Manhattan almost double the total price in Brooklyn
```

```python
#Which top 5 hosts have the highest total price?
air_bnb_total_price = air_bnb.groupby(['host_id', 'host_name', 'calculated_host_listings_count'], as_index=False)['price'].sum()
air_bnb_total_price.sort_values('price', ascending=False).head()

# Based on the total price/night between all the listings a single host has data shows that Sonder is leading with
# a total of $82795 followed by Blueground, Sally, Red Awning, and Kara.
```

```python
# Who currently has no (zero) availability with a review count of 100 or more?
zero_availability = air_bnb.loc[(air_bnb.availability_365==0) & (air_bnb.number_of_reviews>=100)]
zero_availability.count()
zero_availability.get(key=['host_name', "availability_365", 'number_of_reviews' ])
# There are 162 entries that have zero availability and more than 100 reviews.
```

```python
# What host has the highest total of prices and where are they located?
highest_total_prices = air_bnb.groupby(['host_id', 'host_name', 'neighbourhood_group'], as_index=False)['price'].sum()
highest_total_prices.sort_values('price', ascending=False).head(1)

# Sonder has the highest total of prices and is located in Manhattan
```

```python
# When did Danielle from Queens last receive a review?
hey_danielle = air_bnb.loc[(air_bnb.host_name=='Danielle') & (air_bnb.neighbourhood_group=='Queens')]
hey_danielle.sort_values(by=['last_review', 'host_id'], ascending=False).groupby('host_id').head()


# There are 4 different Danielles in Queens. However the latest review for a Danielle in our database was for host_id 26432133 
# and it was posted on 2019-07-08. Danielle from Queend/Long Island City got a review on 2019-06-20,
# Danielle from Queens/Astoria got her latest review on 2018-01-02 and the last Danielle in our database 
# did not get any reviews.
```

## Further Questions

1. Which host has the most listings?

```python
highest_total_prices = air_bnb.groupby(['host_id', 'host_name'], as_index=False)['calculated_host_listings_count'].count()
highest_total_prices.sort_values('calculated_host_listings_count', ascending=False).head(2)

# Sonder has a calculated host listings count of 327 
```

2. How many listings have completely open availability?

```python
complete_availability= air_bnb.loc[(air_bnb.availability_365==365)]
complete_availability.count()

# there are 1295 entries that have a 365 days availability
```

3. What room_types have the highest review numbers?

```python
reviews_by_roomtype = air_bnb.groupby(['room_type'], as_index=False)['number_of_reviews'].sum()
reviews_by_roomtype.sort_values('number_of_reviews', ascending=False)
# The entire home/apt has the highest review numbers and is followed by the private room type.
```
